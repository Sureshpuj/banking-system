```python
print("------------ Welcome to a Banking System! ------------------")
print("1: Create Account")
print("2: My Account (Already Registered)")

choice = input(": ")

if choice == '1':  # Creating an account
    try:
        name = input("\nEnter Account Holder Name: ")
        uid = input("\nEnter UID Number: ")
        
        if not uid.isdigit():
            raise ValueError("UID must contain only numbers.")
        
        account_number = ''
        for i in range(1, 13):
            number = (int(uid) * i) % 10  # Modulus operation
            account_number += str(number)
        
        print(f"\nAccount successfully created!\nNote your account number: {account_number}")

        mobile_number = input("\nEnter Your Mobile Number: ")
        
        if not mobile_number.isdigit() or len(mobile_number) != 10:
            raise ValueError("Invalid mobile number. Enter a 10-digit number.")

        otp = ''
        for i in range(1, 5):
            digit = (int(mobile_number) * i * i) % 10
            otp += str(digit)

        print(f"\nYour OTP is: {otp}")

        entered_otp = input("\nEnter your OTP: ")
        
        if entered_otp == otp:
            print("\nOTP verified successfully!")
            
            while True:
                try:
                    balance = float(input("\nEnter Initial Deposit Amount: "))
                    if balance < 0:
                        raise ValueError("Deposit amount cannot be negative.")
                    break
                except ValueError as e:
                    print(f"Error: {e}. Please enter a valid amount.")
            
            while True:
                try:
                    print("\n---------------------------------------------")
                    print("1: Show Details")
                    print("2: Deposit Money")
                    print("3: Withdraw Money")
                    print("4: Exit")
                    
                    user_choice = int(input("Enter your choice: "))
                    
                    if user_choice == 1:
                        print("\nAccount Details:")
                        print(f"Account Holder Name: {name}")
                        print(f"UID Number: {uid}")
                        print(f"Account Number: {account_number}")
                        print(f"Total Balance: Rs.{balance}")
                    
                    elif user_choice == 2:
                        try:
                            deposit_amount = float(input("Enter Deposit Amount: "))
                            if deposit_amount <= 0:
                                raise ValueError("Deposit amount must be greater than zero.")
                            balance += deposit_amount
                            print(f"\nA/C {account_number} credited by Rs.{deposit_amount}.\nAvailable Balance: Rs.{balance}")
                        except ValueError as e:
                            print(f"Error: {e}. Please enter a valid amount.")
                    
                    elif user_choice == 3:
                        try:
                            withdraw_amount = float(input("Enter Withdrawal Amount: "))
                            if withdraw_amount <= 0:
                                raise ValueError("Withdrawal amount must be greater than zero.")
                            elif withdraw_amount > balance:
                                print("Insufficient Balance!")
                            else:
                                balance -= withdraw_amount
                                print(f"\nA/C {account_number} debited by Rs.{withdraw_amount}.\nAvailable Balance: Rs.{balance}")
                        except ValueError as e:
                            print(f"Error: {e}. Please enter a valid amount.")
                    
                    elif user_choice == 4:
                        print("Thank you for using the banking system. Goodbye!")
                        break
                    
                    else:
                        print("Invalid choice. Please enter a valid option.")
                
                except ValueError:
                    print("Invalid input. Please enter a number from the menu options.")
        
        else:
            print("Incorrect OTP. Please try again.")
    
    except ValueError as e:
        print(f"Error: {e}. Please try again.")

else:
    print("\nYour Account is Not Found. Please create an account first.")
```
